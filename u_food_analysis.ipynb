## Data Cleaning & Standarizing 

```python
## check for duplicates
food[food.duplicated()].sum()
```

```python
food.drop_duplicates(keep='first', inplace=True)
```

```python
## checking if the duplicates were romoved
food.info()
```

data is standardized not much transformation needed
* no Nulls in the data
* all the data tupes are int64 except for the income


```python
## combine columns 'Kidhome' and 'Teenhome' into one column 'Children'
food['Children'] = food['Kidhome'] + food['Teenhome']
```

## Combine all the Marital status into one column
* divoreced = 5
* married = 4
* single = 3
* together = 2
* widow = 1
 

```python


food['marital_Divorced'] = food['marital_Divorced'].replace({1:5,0:0})
food['marital_Married'] = food['marital_Married'].replace({1:4,0:0})
food['marital_Single'] = food['marital_Single'].replace({1:3,0:0})
food['marital_Together'] = food['marital_Together'].replace({1:2,0:0})
food['marital_Widow'] = food['marital_Widow'].replace({1:1,0:0})
```

```python
## combine all the martial status columns into one column 'Marital_Status'
food['marital_status'] = food['marital_Divorced'] + food['marital_Married'] + food['marital_Single'] + food['marital_Together'] + food['marital_Widow']
```

```python
## string values for the marital status
food['marital_status_str'] = food['marital_status'].map({1:'Widow',2:'Together',3:'Single',4:'Married',5:'Divorced'})
```

## combine education 
* education_2n Cycle = 1
* education_Basic = 2
* education_Graduation = 3 
* education_Master = 4
* education_PhD = 5

```python

food['education_2n Cycle'] = food['education_2n Cycle'].replace({1:1,0:0})
food['education_Basic'] = food['education_Basic'].replace({1:2,0:0})
food['education_Graduation'] = food['education_Graduation'].replace({1:3,0:0})
food['education_Master'] = food['education_Master'].replace({1:4,0:0})
food['education_PhD'] = food['education_PhD'].replace({1:5,0:0})
```

```python
food['education'] = food['education_2n Cycle'] + food['education_Basic'] + food['education_Graduation'] + food['education_Master'] + food['education_PhD']
```

## check for campaigns accepted

```python
food['Accepted_campaign'] = food['AcceptedCmp1'] + food['AcceptedCmp2'] + food['AcceptedCmp3'] + food['AcceptedCmp4'] + food['AcceptedCmp5'] + food['Response']
```

```python
##  'Accepted_campaign' cloumn turned into yes or no 

food['Accepted_campaign'] = (food['Accepted_campaign'] > 0).astype(int)
```

```python
food = food.drop('marital_status_str', axis=1)
```

```python
age_group = ['<30', '31-40', '41-50', '51-60', '61-70', '71-85']
def assign_age_group(age):
    if age < 30:
        return '<30'
    elif age >= 30 and age < 41:
        return '31-40'
    elif age >= 41 and age < 51:
        return '41-50'
    elif age >= 51 and age < 61:
        return '51-60'
    elif age >= 61 and age < 71:
        return '61-70'
  
    else:
        return '71-85'
    
food['Age_group'] = food['Age'].apply(assign_age_group)
```
